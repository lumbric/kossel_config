```python
def g29out_to_array(tty_strout):
    """Convert auto-level output to numpy.array."""
    regex = re.compile(r'Bed x: (-?\d+.\d+) y: (-?\d+.\d+) z: (-?\d+.\d+)', re.MULTILINE)

    # TODO loop might not be very efficient
    val_list = []
    for line in tty_strout.splitlines():
        res = regex.match(line)
        if res is None:
            continue
        vals = res.groups()
        val_list.append([float(val) for val in vals])
    return np.array(val_list)

def plot_bed3d(xyz, resolution=7):
    """Pass measure points xyz in shape (n,3)."""
    x,y,z = xyz.T
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    # there is no plot_trisurf() with old matplotlib, but this does not work well because of NaNs and inf...
    #_x = np.linspace(x.min(), x.max(), RESOLUTION)
    #_y = np.linspace(y.min(), y.max(), RESOLUTION)
    #x_grid, y_grid = np.meshgrid(_x, _y)
    #z_interp = griddata((x, y), z, (x_grid, y_grid), method='linear')
    #surf = ax.plot_surface(x_grid, y_grid, z_interp)
    
    ax.plot_trisurf(x, y, z)
    ax.scatter(x, y, zs=z, color='r')

    return fig
```
